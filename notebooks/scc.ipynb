```python
import networkx as nx
import os
import re
import matplotlib.pyplot as plt
import numpy as np
```

```python
file_path = "../csv/net"
imgs_path = "../img"
regex = re.compile("gephi.*csv")
net_files = list(filter(regex.match, sorted(os.listdir(file_path))))
net_files
```

```python
'''Network 0'''
net_0 = nx.DiGraph()
with open(f"{file_path}/{net_files[0]}") as f:
    file_content = (f.read().splitlines())

for line in file_content:
    nodes = [int(x) for x in line.split(',')]
    edges = [(nodes[0],x) for x in nodes[1:]]
    if len(edges) > 1:
        net_0.add_edges_from(edges)
    else:
        net_0.add_node(nodes[0])

nx.draw_circular(net_0, with_labels=True)
```

```python
'''Network 0 SCC'''
net_0_comps = nx.strongly_connected_components(net_0)
print("Number of nodes in SCC", max([len(c) for c in sorted(net_0_comps, key=len, reverse=True)]))

net_0_comps = list(sorted(nx.strongly_connected_components(net_0), key=len, reverse=True)[0])
net_0_scc_subgraph = net_0.subgraph(net_0_comps)
print("Average shortest path length", nx.average_shortest_path_length(net_0_scc_subgraph))
```
